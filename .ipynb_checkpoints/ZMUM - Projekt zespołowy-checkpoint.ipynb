## Zadanie 4
Napisz metodę regresji krokowej w tył i za jej pomocą wybierz 5 najbardziej wartościowych cech.

```python
def create_regularization_matrix(st):
    m = np.eye(st)
    m[0,0] = 0
    return m

def regresja_L2(X, y, lamb):
    lamb_m = lamb * create_regularization_matrix(X.shape[1])
    return np.linalg.inv(X.T.dot(X) + lamb_m).dot(X.T).dot(y)
```

```python
def backward_regress(X_train, y_train, lambdas):
    scores_dict = {}
    scores_dict['lamb'] = lambdas
    scores_dict['coef'] = []
    for lamb in lambdas:
        w = regresja_L2(X_train, y_train, lamb)
        scores_dict['coef'].append(w)
    return scores_dict
```

```python
all_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated',
       'sqft_living15', 'sqft_lot15', 'bed_bath_rooms',
       'rating_compound', 'sqft_lot_log', 'sqft_living_lot',
       'sqft_above_ratio']

lambdas = [1*(2**i) for i in range(50)]
scores_df = pd.DataFrame(backward_regress(domy_train[all_features], ceny_train, lambdas))
```

## Zadanie 5
Dokonaj normalizacji cech (wszystkie podane + 5 zaproponowanych w Zadaniu 1) i przy pomocy metody Lasso lub LARS wybierz 5 najbardziej wartościowych cech.

!!! Przy przekształceniach unikaj korzystania z informacji zawartej w zbiorze walidacyjnym i testowym !!!

```python
all_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built',
       'sqft_living15', 'sqft_lot15', 'bed_bath_rooms',
       'rating_compound', 'sqft_lot_log', 'sqft_living_lot',
       'sqft_above_ratio']
```

```python
norm_features = []
for feat in all_features:
    norm_features.append('norm_' + feat)
    norm = np.linalg.norm(domy[feat])
    domy['norm_' + feat] = domy[feat] / norm
    domy_train['norm_' + feat] = domy_train[feat] / norm
    domy_test['norm_' + feat] = domy_test[feat] / norm    
```

```python
class Lasso:
    def __init__(self, alpha: float = 1.0, max_iter: int = 1000, fit_intercept: bool = True) -> None:
        self.alpha: float = alpha
        self.max_iter: int = max_iter
        self.fit_intercept: bool = fit_intercept 
        self.coef_ = None
        self.intercept_ = None

    def _soft_thresholding_operator(self, x: float, lambda_: float) -> float:
        if x > 0.0 and lambda_ < abs(x):
            return x - lambda_
        elif x < 0.0 and lambda_ < abs(x):
            return x + lambda_
        else:
            return 0.0

    def fit(self, X: np.ndarray, y: np.ndarray):
        if self.fit_intercept:
            X = np.column_stack((np.ones(len(X)),X))

        beta = np.zeros(X.shape[1])
        if self.fit_intercept:
            beta[0] = np.sum(y - np.dot(X[:, 1:], beta[1:]))/(X.shape[0])
        
        for iteration in range(self.max_iter):
            start = 1 if self.fit_intercept else 0
            for j in range(start, len(beta)):
                tmp_beta = beta.copy()
                tmp_beta[j] = 0.0
                r_j = y - np.dot(X, tmp_beta)
                arg1 = np.dot(X[:, j], r_j)
                arg2 = self.alpha*X.shape[0]

                beta[j] = self._soft_thresholding_operator(arg1, arg2)/(X[:, j]**2).sum()

                if self.fit_intercept:
                    beta[0] = np.sum(y - np.dot(X[:, 1:], beta[1:]))/(X.shape[0])

        if self.fit_intercept:
            self.intercept_ = beta[0]
            self.coef_ = beta[1:]
        else:
            self.coef_ = beta

        return self

    def predict(self, X: np.ndarray):
        y = np.dot(X, self.coef_)
        if self.fit_intercept:
            y += self.intercept_*np.ones(len(y))
        return y
```

```python
def lasso_feature_picker(X_train, y_train, lambdas):
    scores_dict = {}
    scores_dict['lamb'] = lambdas
    scores_dict['coef'] = []
    for lamb in lambdas:
        model = Lasso(lamb)
        model.fit(X_train, y_train)
        scores_dict['coef'].append(model.coef_)
    return scores_dict
```

```python
lambdas = [1*(2**i) for i in range(6)]
lambdas.extend([i for i in range(48, 129, 16)])
print('lambdas', lambdas)
scores_df = pd.DataFrame(lasso_feature_picker(domy_train[norm_features], ceny_train, lambdas))
```

## Zadanie 6
Dla 5 wybranych w Zadaniu 5 cech dokonaj doboru współczynnika regularyzacji dla metody grzbietowej. 

```python
def create_regularization_matrix(st):
    m = np.eye(st)
    m[0,0] = 0
    return m

def regresja_ridge(X, y, lamb):
    lamb_m = lamb * create_regularization_matrix(X.shape[1])
    return np.linalg.inv(X.T.dot(X) + lamb_m).dot(X.T).dot(y)
```

```python
def ridge_scores(X_train, y_train, X_test, y_test, lambdas):
    scores_dict = {}
    scores_dict['lamb'] = lambdas
    scores_dict['coef'] = []
    scores_dict['train_score'] = []
    scores_dict['test_score'] = []
    for lamb in lambdas:
        w = regresja_L2(X_train, y_train, lamb)
        scores_dict['coef'].append(w)
        scores_dict['train_score'].append(np.mean((X_train.dot(w) - y_train)**2))
        scores_dict['test_score'].append(np.mean((X_test.dot(w) - y_test)**2))
    return scores_dict
```

```python
lambdas = [1e-12*(2**i) for i in range(45)]
# lambdas.extend([i for i in range(2,11)])
# lambdas.extend([i for i in range(12,21,2)])
# lambdas.extend([i for i in range(25,51,5)])
scores_df = pd.DataFrame(ridge_scores(domy_train[chosen_ex_5], ceny_train, domy_test[chosen_ex_5], ceny_test, lambdas))
```
